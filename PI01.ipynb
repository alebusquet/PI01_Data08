```python
# Generar campo id: Cada id se compondrá de la primera letra del nombre de la plataforma, seguido del show_id 
# ya presente en los datasets (ejemplo para títulos de Amazon = as123)

for i in range(len(df_unico)):
    palabra = df_unico.at[i, "platform"]
    letra = palabra[0]
    df_unico.at[i, "show_id"] = letra + df_unico.at[i, "show_id"]
```

```python
# Veo como quedo
df_unico.head()
```

```python
# Los valores nulos del campo rating deberán reemplazarse por el string “G” (corresponde al maturity 
# rating: “general for all audiences”).
df_unico["rating"].fillna("G", inplace=True)


# Con firmo que no quedaron valores nulos.
df_unico["rating"].isnull().value_counts()
```

```python
# De haber fechas, deberán tener el formato AAAA-mm-dd.
df_unico['date_added'] = pd.to_datetime(df_unico['date_added'], infer_datetime_format=True)
df_unico.head()
```

```python
# Los campos de texto deberán estar en minúsculas, sin excepciones.
# Me parecio no tener sentido armar una funcion con una linea, por lo que ejecuto las columnas de a una.
df_unico["show_id"] = df_unico["show_id"].str.lower()
df_unico["type"] = df_unico["type"].str.lower()
df_unico["title"] = df_unico["title"].str.lower()
df_unico["cast"] = df_unico["cast"].str.lower()
df_unico["rating"] = df_unico["rating"].str.lower()
df_unico["listed_in"] = df_unico["listed_in"].str.lower()
df_unico["description"] = df_unico["description"].str.lower()
df_unico["platform"] = df_unico["platform"].str.lower()

df_unico.head()
```

```python
# La columna "duration" debe convertirse en dos: "duration_int" y "duration_type". La primera debera ser un integer 
# y la segunda un string, indicando la unidad de medición de duración: min (minutos) o season (temporadas).


# Completo los nulos de la columna con 0 para que despues no me devuelvan error.
df_unico["duration"].fillna("0", inplace=True)


# Genero las 2 columnas requeridas.
df_unico.insert(loc = 8, column = "duration_int", value=0)
df_unico.insert(loc = 9, column = "duration_type", value="sin dato")


# Realizo los cambios: los numeros van a una columna (que quedara como entero) y las palabras van a otra.
for i in range(len(df_unico)):
    if df_unico.at[i, "duration"] != "0":
        campo = df_unico.at[i, "duration"]
        div = campo.split(" ")
        divint = int(div[0])
        
        df_unico.at[i, "duration_int"] = divint
        df_unico.at[i, "duration_type"] = div[1]


# Una vez finalizado y considerando que no hubo errores, elimino la columna original.
df_unico.drop(["duration"], axis=1, inplace=True)
```

#### Preprocesamiento de los datos

```python
# --> Trabajo con los actores.

# Reemplazo los datos faltantes de la columna para evitar errores posteriores.
df_unico["cast"].fillna("sin dato", inplace=True)


# Genero una nueva columna de trabajo.
df_unico.insert(loc = 4, column = "cast1", value="sin dato")


# Paso a los actores a listas, que luego las reinserto en el dataframe.
df_unico["cast1"] = df_unico["cast"].str.split("," or ":" or "|")
df_unico.drop(columns=["cast"],inplace=True)
df_unico.rename(columns={"cast1":"cast"},inplace=True)


# Chequo las listas en columna "cast".
df_unico.head
```

```python
# Creación de DataFrame para los actores, a fin de utilizarlo en las consultas.

# Primero inicializo un diccionario.
dix_actors = {"actor":[], 
              "año":[],
              "plataforma":[]}


# Voy recorriendo los actores y voy llevando su info (año y plataforma) al diccionario
for index, actores in enumerate(df_unico["cast"]):
    for actor in actores:
        if actor != "sin dato":           
            dix_actors["actor"].append(actor)
            dix_actors["año"].append(df_unico["release_year"][index])
            dix_actors["plataforma"].append(df_unico["platform"][index])


# Genero el dataframe en base al diccionario
df_actors = pd.DataFrame(dix_actors)


# Mirada al dataframe
df_actors.head(10)
```

Película con mayor duración con filtros opcionales de AÑO, PLATAFORMA Y TIPO DE DURACIÓN. <br>
La función debe llamarse get_max_duration(year, platform, duration_type).

```python
# Armo la funcion considerando que los parametros son opcionales.
#def get_max_duration(year: Union[int, None] = 0, platform: Union[str, None] = None, duration_type: Union[str, None] = None):
def get_max_duration(year: int, platform: str, duration_type:str):

    if (year != 0) & (platform != None) & (duration_type != None):
        data = df_unico[(df_unico["release_year"] == year) & (df_unico["platform"] == platform) & (df_unico["duration_type"] == duration_type)]["duration_int"].max()
    elif (year == 0) & (platform != "") & (duration_type != None):
        data = df_unico[(df_unico["platform"] == platform) & (df_unico["duration_type"] == duration_type)]["duration_int"].max()
    elif (year == 0) & (platform == "") & (duration_type != None):
        data = df_unico[(df_unico["duration_type"] == duration_type)]["duration_int"].max()
        platform = "Sin especificar"
    elif (year == 0) & (platform != None) & (duration_type == None):
        data = df_unico[(df_unico["platform"] == platform)]["duration_int"].max()
        duration_type = "Sin especificar"
    else:
        data = df_unico["duration_int"].max()
        platform = "Sin especificar"
        duration_type = "Sin especificar"

    message = f"Máximo para plataforma: {platform}, para el año: {year} en tipo: {duration_type} es"
    return {message:data}
```

```python
# Pruebo la funcion.
get_max_duration(0, "netflix", "min")
```

```python
# Cargo los dataframes.
df1 = pd.read_csv("./Datasets/ratings/1.csv")
df2 = pd.read_csv("./Datasets/ratings/2.csv")
df3 = pd.read_csv("./Datasets/ratings/3.csv")
df4 = pd.read_csv("./Datasets/ratings/4.csv")
df5 = pd.read_csv("./Datasets/ratings/5.csv")
df6 = pd.read_csv("./Datasets/ratings/6.csv")
df7 = pd.read_csv("./Datasets/ratings/7.csv")
df8 = pd.read_csv("./Datasets/ratings/8.csv")


# Los reuno en uno solo.
df_concat = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8], axis=0)
```

```python
# Dejo solo un registro por pelicula, con el promedio de sus puntajes.
df_promedio = df_concat.groupby("movieId").agg({"userId": "first", "rating": "mean", "timestamp": "first"}).reset_index()
```

```python
# Renombro la columna de "rating" para que defina mejor su contenido, y la de "movieId", esta ultima para que se llame igual que en df_unico.
df_promedio = df_promedio.rename(columns={"rating": "rating_prom", "movieId": "show_id"})
```

```python
# Elimino las columnas "userId" y "timestamp" que  no las voy a necesitar para esta funcion.
df_promedio.drop(["userId", "timestamp"], axis=1, inplace=True)
```

```python
# Creo un dataframe que unifique los datos de la pelicula, la plataforma, el año y el puntaje, que ahora se encuentran en 2 dataframes distintos.
df_completo = pd.merge(df_unico, df_promedio, on='show_id')
```

```python
df_completo
```

```python
# Presento la funcion.
def get_score_count(platform:str, scored:int, year:int):
    df_filtrado = df_completo.loc[(df_completo['platform'] == platform) & (df_completo['release_year'] == year)]
    df_filtrado = df_filtrado.loc[df_filtrado['rating_prom'] > scored]
    cantidad_pelis = len(df_filtrado)
    
    message = f"Cantidad de películas en {platform}, con un puntaje mayor a {scored}, para el año {year} es"
    return {message: cantidad_pelis}
```

```python
# Realizo una consulta.
get_score_count("amazon", 3, 2021)
```

```python
# Defino la funcion.
def get_actor(plataforma:str, año:int):
    data = df_actors[(df_actors["año"] == año) & (df_actors["plataforma"] == plataforma)]["actor"]
    message = f"Actor que mas se repite en {plataforma} en el año {año}"
    return {message:data.value_counts().index[0]}
```
